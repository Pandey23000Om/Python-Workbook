```python
import pandas as pd
```

### Read in data

```python
df = pd.read_csv('pokemon_pokedex_data.csv')
```

### Take care of null values

```python
df.isnull()
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df = df.fillna('NA')
```

### Change data type of a column

```python
df['pokedex_id']=df['pokedex_id'].astype(str)
```

```python
df.dtypes
```

### Data aggregating and grouping

```python
grouped_mean = df.groupby('type_1')['total_attack'].mean().sort_values()
print(grouped_mean)
```

```python
grouped_mean = df.groupby('type_1')['defense'].mean().sort_values()
print(grouped_mean)
```

```python
grouped_mean2 = df.groupby(['type_1','type_2'])['attack'].mean().sort_values()
print(grouped_mean2)
grouped_mean2.to_json("grouped_mean.json")
```
